### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_top_max = df_cities.nlargest(5,"population_total")
df_top_max
```

```python
plt.Figure()
x= df_top_max.city
y= df_top_max.population_total/1000
colores = ["mediumslateblue", "silver", "lightcoral", "olive", "rosybrown"]

plt.bar(x,y, color = colores)
plt.title("Ciudades mas pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.ylim((0,5000))

for i in range(len(df_top_max)):
    plt.text(i, y.iloc[i] + 75, y.iloc[i], ha = 'center');
```

```python
plt.Figure()
plt.hist(df_cities.area_total_km2, bins = 50, label = "Hist area")
plt.xlabel("Superficie (Km$^2$)")
plt.ylabel("Numero de ciudades")
plt.legend()
```

```python
x= df_cities.area_water_km2
y = df_cities.population_total/1000
```

```python
plt.Figure()
plt.scatter(x, y, color = "green", alpha = 0.5)
plt.xlabel("Area de agua en la ciudad (Km$^2$)")
plt.ylabel("Pblacion (miles de habitantes)")
plt.title("Poblacion por kilometro cuadrado de agua")
# Aqu los annotate
coordenadas = (88,3884)
plt.annotate("Outliers", xy =coordenadas, textcoords='offset points', xytext=(-50, -50),
             arrowprops = dict(arrowstyle="->", connectionstyle="arc3")
            )
coordenadas = (122,1345)
plt.annotate("", xy =coordenadas, textcoords='offset points', xytext=(-120, 100),
             arrowprops = dict(arrowstyle="->", connectionstyle="arc3")
            );
```

```python
df_cities.nlargest(2,"area_water_km2")
```

```python
plt.Figure()
plt.boxplot(df_cities.elevation_m, orientation = "horizontal", patch_artist = True, boxprops=dict(edgecolor='black', facecolor = "red"), medianprops=dict(color='black'))
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California");
```
