### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_top_max = df_cities.nlargest(5,"population_total")
df_top_max
```

```python
x= df_top_max.city
y= df_top_max.population_total/1000
colores = ["blue", "gray", "red", "olive", "brown"]

plt.Figure(figsize = (8,5))
plt.bar(x,y, color = colores, alpha = 0.5)
plt.title("Ciudades mas pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.ylim((0,5000))

for i in range(len(df_top_max)):
    plt.text(i, y.iloc[i] + 75, y.iloc[i], ha = 'center')

plt.show()
```

```python
#COnsultamos si la familia de fuentes está instalada
import matplotlib

print(matplotlib.font_manager.findfont("Bradley Hand ITC"))
```

```python
#Si da problemas ejecutar, entrar a la carpeta y eliminar el archivo json
print(matplotlib.get_cachedir())
```

```python
plt.Figure(figsize = (8,5))
plt.hist(df_cities.area_total_km2, 
         bins = 50, 
         label = "Hist area", 
         alpha = 0.7)
plt.xlabel("Superficie (Km$^2$)", fontdict = dict(family = "Bradley Hand ITC", size = 18))
plt.ylabel("Numero de ciudades", fontdict = dict(family = "Bradley Hand ITC", size = 18))
plt.legend()

plt.show()
```

SOLUCION  FIGURA 3

```python
x= df_cities.area_water_km2
y = df_cities.population_total/1000
```

```python
plt.Figure(figsize = (8,5))
plt.scatter(x, y, color = "green", alpha = 0.5)
plt.xlabel("Area de agua en la ciudad (Km$^2$)")
plt.ylabel("Pblacion (miles de habitantes)")
plt.title("Poblacion por kilometro cuadrado de agua")

# Aquí los annotate
coordenadas = (88,3884)
plt.annotate("Outliers", xy =coordenadas, xytext=(60,3500),
             arrowprops = dict(arrowstyle="->"))

coordenadas = (122,1345)
plt.annotate("        ", xy =coordenadas, xytext=(60,3500),
             arrowprops = dict(arrowstyle="->"))

plt.show()
```

SOLUCION FIGURA 4

```python
plt.Figure(figsize = (8,5))
plt.boxplot(df_cities.elevation_m, orientation = "horizontal", patch_artist = True, boxprops=dict(edgecolor='black', facecolor = "red"), medianprops=dict(color='black'))
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.yticks([])

plt.show()
```

```python
fig = plt.figure(figsize = (15,15))
plt.subplots_adjust(hspace = 0.2, wspace = 0.2)


plt.subplot(2,2,1)
x= df_top_max.city
y= df_top_max.population_total/1000
colores = ["mediumslateblue", "silver", "lightcoral", "olive", "rosybrown"]
plt.bar(x,y, color = colores)
plt.title("Ciudades mas pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.ylim((0,5000))
for i in range(len(df_top_max)):
    plt.text(i, y.iloc[i] + 75, y.iloc[i], ha = 'center');

    
plt.subplot(2,2,2)
plt.hist(df_cities.area_total_km2, bins = 50, label = "Hist area")
plt.xlabel("Superficie (Km$^2$)")
plt.ylabel("Numero de ciudades")
plt.legend()


plt.subplot(2,2,3)
x= df_cities.area_water_km2
y = df_cities.population_total/1000
plt.scatter(x, y, color = "green", alpha = 0.5)
plt.xlabel("Area de agua en la ciudad (Km$^2$)")
plt.ylabel("Pblacion (miles de habitantes)")
plt.title("Poblacion por kilometro cuadrado de agua")
# Aquí los annotate
coordenadas = (88,3884)
plt.annotate("Outliers", xy =coordenadas, xytext=(60,3500),
             arrowprops = dict(arrowstyle="->"))
coordenadas = (122,1345)
plt.annotate("        ", xy =coordenadas, xytext=(60,3500),
             arrowprops = dict(arrowstyle="->"))



plt.subplot(2,2,4)
plt.boxplot(df_cities.elevation_m, orientation = "horizontal", patch_artist = True, boxprops=dict(edgecolor='black', facecolor = "red"), medianprops=dict(color='black'))
plt.xlabel("Altitud (m)");
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.yticks([])

plt.show()
```
